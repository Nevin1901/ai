```python
import numpy as np
from sklearn import linear_model, datasets, metrics
import matplotlib.pyplot as plt
```

```python
(new_x, new_y) = datasets.load_diabetes(return_X_y=True)
print(new_x[:, None, 2]) # all the None, 2 does after getting the first column is puts each value into an array of its own
```

```python
(data_x, data_Y) = datasets.load_diabetes(return_X_y=True)
data_x = data_x[:, None, 2] # use one feature, gotten from example
weights = np.ones(len(data_Y))
Y = np.stack((data_Y, weights), axis=1)

# split into train and test sets
split_x = np.split(data_x, 2)
split_y = np.split(Y, 2)

train_x = split_x[0]
test_x = split_x[1]

train_Y = split_y[0]
test_Y = split_y[1]
```

```python
model = linear_model.LinearRegression()
model.fit(train_x, train_Y)
```

```python
y_pred = model.predict(test_x)
print("Mean squared error: ", metrics.mean_squared_error(test_Y, y_pred))
print("r2 Score: ", metrics.r2_score(test_Y, y_pred))

# graph the data

plt.scatter(test_x, test_Y[:, 0])
plt.plot(test_x, y_pred[:, 0])
```
